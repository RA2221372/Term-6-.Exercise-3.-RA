```python
# TODO: تابع نمایش تصاویر رندوم را کامل کنید.

import os
import random
import matplotlib.pyplot as plt
from PIL import Image

def visualize_two_sample_images_smart(folder_path, title="Sample Images"):
    """
    دو تصویر تصادفی از پوشه یا زیرپوشه‌هایش بارگذاری و نمایش می‌دهد.
    فقط فایل‌های تصویری (.jpg, .jpeg, .png) شمارش و نمایش داده می‌شوند.
    """
    valid_exts = (".jpg", ".jpeg", ".png")
    
    # بررسی وجود پوشه
    if not os.path.exists(folder_path):
        print(f" پوشه پیدا نشد: {folder_path}")
        return
    
    items = os.listdir(folder_path)
    
    # اگر فقط یک زیرپوشه داشت، برو داخلش
    if len(items) == 1 and os.path.isdir(os.path.join(folder_path, items[0])):
        folder_path = os.path.join(folder_path, items[0])
        items = os.listdir(folder_path)
    
    # گرفتن فقط فایل‌های تصویری
    images = [f for f in items if f.lower().endswith(valid_exts)]
    
    if len(images) == 0:
        print(f" هیچ تصویری در پوشه {folder_path} پیدا نشد!")
        return
    
    # انتخاب دو تصویر رندوم
    sample_images = random.sample(images, min(2, len(images)))
    
    # نمایش تصاویر
    plt.figure(figsize=(6, 3))
    for i, img_name in enumerate(sample_images):
        img_path = os.path.join(folder_path, img_name)
        try:
            img = Image.open(img_path).convert("RGB").resize((150, 150))
            plt.subplot(1, 2, i+1)
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"{title} #{i+1}")
        except Exception as e:
            print(f"⚠️ خطا در بارگذاری {img_name}: {e}")
    
    plt.tight_layout()
    plt.show()

# مسیرها
BASE_DIR = r"C:\Users\Mahdi\Downloads\rps_dataset-20250830T081629Z-1-001\rps_dataset"
PAPER_DIR = os.path.join(BASE_DIR, "paper")
ROCK_DIR = os.path.join(BASE_DIR, "rock")
SCISSORS_DIR = os.path.join(BASE_DIR, "scissors")

# نمایش تصاویر نمونه
visualize_two_sample_images_smart(PAPER_DIR, title="Paper")
visualize_two_sample_images_smart(ROCK_DIR, title="Rock")
visualize_two_sample_images_smart(SCISSORS_DIR, title="Scissors")


```

* [Load & Preprocess Images | Tensorflow](https://www.tensorflow.org/tutorials/load_data/images)
* [image_dataset_from_directory | Tensorflow](https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image_dataset_from_directory)
* [How to Use](https://datascience.stackexchange.com/questions/87469/how-to-load-all-images-using-image-dataset-from-directory-function)

```python
# TODO: اندازه تصاویر و اندازه هر بچ (تعداد تصاویر در هر دسته آموزش) را مشخص کنید
import tensorflow as tf

IMG_SIZE = (150, 150)   # همه تصاویر به این اندازه تغییر سایز داده می‌شوند
BATCH_SIZE = 32          # تعداد تصاویر در هر دسته (batch) هنگام آموزش

# TODO: داده‌های آموزش را بارگذاری کنید
train_dataset = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,                  # مسیر دیتاست
    validation_split=0.2,      # 20٪ داده برای اعتبارسنجی
    subset="training",         # این مجموعه، آموزش است
    seed=123,                  # برای reproducibility
    image_size=IMG_SIZE,       # تغییر سایز تصاویر
    batch_size=BATCH_SIZE       # تعداد تصاویر در هر batch
)

# TODO: داده‌های اعتبارسنجی را بارگذاری کنید
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,      # همان 20٪ برای اعتبارسنجی
    subset="validation",       # این مجموعه، اعتبارسنجی است
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
```

    <li>
        در این مرحله باید داده‌های آموزش و اعتبارسنجی را برای آموزش بهینه مدل آماده‌سازی کنید.
        برای این کار:
        <li>        از .cache() برای ذخیره‌سازی داده‌ها در حافظه استفاده کنید تا خواندن آن‌ها در هر epoch سریع‌تر انجام شود.</li>
        <li>        از .shuffle() برای ایجاد تنوع در ترتیب داده‌های آموزشی استفاده کنید تا مدل دچار overfitting نشود.</li>
        <li>        از .prefetch() برای بارگذاری داده‌ها به‌صورت غیرهمزمان استفاده کنید تا بهره‌وری پردازنده/کارت گرافیک افزایش یابد.</li>
    </li>

```python
# TODO : کش کردن داده‌ها در حافظه برای افزایش سرعت آموزش (فقط در صورت کافی بودن حافظه RAM)

AUTOTUNE = tf.data.AUTOTUNE  # برای prefetch خودکار و بهینه‌سازی

# --- داده‌های آموزش ---
train_dataset = (
    train_dataset
    .cache()   # کش کردن داده‌ها در حافظه برای افزایش سرعت آموزش
    .shuffle(1000)  # shuffle برای جلوگیری از overfitting
    .prefetch(buffer_size=AUTOTUNE)  # بارگذاری غیرهمزمان
)

# --- داده‌های اعتبارسنجی ---
validation_dataset = (
    validation_dataset
    .cache()   # کش کردن داده‌ها در حافظه
    .prefetch(buffer_size=AUTOTUNE)
)
```

<div dir='rtl'>
  <font size=4>
    <li>
        در این بخش، شما باید یک لایه افزایش داده (Data Augmentation) طراحی کنید که تصاویر را به‌صورت تصادفی بچرخاند و قرینه کند، سپس تعدادی از تصاویر تقویت‌شده را برای بررسی بصری نمایش دهید.
      هدف این مرحله افزایش تنوع داده‌های آموزشی برای بهبود تعمیم‌پذیری مدل است.
    </li>
</font>

* [Data Augmentation | Tensorflow](https://www.tensorflow.org/tutorials/images/data_augmentation)
* [Data Augmentation | Example](https://pyimagesearch.com/2021/06/28/data-augmentation-with-tf-data-and-tensorflow/)

```python
# TODO: افزایش داده

from tensorflow.keras import layers
import matplotlib.pyplot as plt

# --- طراحی لایه Data Augmentation ---
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),      # قرینه کردن افقی تصاویر
    layers.RandomRotation(0.2),           # چرخش تصادفی تصاویر تا ±20%
    layers.RandomZoom(0.1)                # بزرگ‌نمایی تصادفی ±10%
])

# --- نمایش چند تصویر نمونه پس از Data Augmentation ---
for images, labels in train_dataset.take(1):
    plt.figure(figsize=(10, 5))
    
    for i in range(6):
        augmented_image = data_augmentation(tf.expand_dims(images[i], 0))  # افزودن بعد batch
        plt.subplot(2, 3, i+1)
        plt.imshow(augmented_image[0].numpy().astype("uint8"))
        plt.axis("off")
    
    plt.suptitle("نمونه تصاویر تقویت‌شده")
    plt.show()
```

         در اینجا هدف اصلی نرمال‌سازی مقادیر پیکسل‌ها است، به این صورت که مقادیر اصلی که در بازه [0, 255] هستند، به بازه [0.0, 1.0] تبدیل می‌شوند.
         یک تابع تعریف شده که هر تصویر را از نوع int به float تبدیل می‌کند و سپس آن را بر ۲۵۵ تقسیم می‌کند تا نرمال شود. این تابع را کامل کنید و سپس  روی کل داده‌های آموزش و اعتبارسنجی با استفاده از متد map اعمال شده است.

```python
import tensorflow as tf

# 1️تنظیم پارامترها
IMG_SIZE = (150, 150)   # اندازه تصویر
BATCH_SIZE = 32         # تعداد تصاویر در هر batch
BASE_DIR = r"C:\Users\Mahdi\Downloads\rps_dataset-20250830T081629Z-1-001\rps_dataset"  # مسیر دیتاست

# 2️ ساخت دیتاست آموزش
train_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,    # 80% آموزش، 20% اعتبارسنجی
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# 3️ ساخت دیتاست اعتبارسنجی
val_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

#  چک کردن
print("دیتاست آموزش:", train_ds)
print("دیتاست اعتبارسنجی:", val_ds)
```

```python
# تعریف تابع پیش‌پردازش شامل نرمال‌سازی مقادیر پیکسل
def preprocess(image, label):
 image = tf.cast(image, tf.float32) / 255.0  
 return image, label
# اعمال پیش‌پردازش روی دیتاست‌ها
train_ds = train_ds.map(preprocess)
val_ds = val_ds.map(preprocess)
```

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر

import tensorflow as tf
from tensorflow.keras import layers, models

# لایه افزایش داده (Data Augmentation) قبلاً تعریف شده
# data_augmentation = tf.keras.Sequential([...])

IMG_SIZE = (150, 150)

model = models.Sequential([
    layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3)),  # لایه ورودی جداگانه
    data_augmentation,                                  # اعمال افزایش داده
    layers.Rescaling(1./255),                           # نرمال‌سازی پیکسل‌ها
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(3, activation='softmax')  # سه کلاس: Rock, Paper, Scissors
])
```

```python
# TODO: مدل را کامپایل کنید

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # چون label ها integer هستند
    metrics=['accuracy']
)

model.summary()
```

```python
# TODO: مدل را آموزش دهید
from tensorflow.keras.callbacks import EarlyStopping

# تعریف callback برای توقف زودهنگام
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=3,       # اگر ۳ بار پشت سر هم بهبود نداشت متوقف میشه
    restore_best_weights=True
)

# آموزش مدل
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,              
    callbacks=[early_stopping]
)
```

```python
def plot_training_history(history):
    """
    رسم گراف‌های دقت و خطا برای داده‌های آموزش و اعتبارسنجی.

    این تابع دو نمودار را رسم می‌کند:
    1. نمودار دقت (Accuracy) که دقت مدل در طول آموزش و اعتبارسنجی را نمایش می‌دهد.
    2. نمودار خطا (Loss) که میزان خطای مدل را در طول آموزش و اعتبارسنجی نشان می‌دهد.

    پارامترها:
    - history: شی‌ای که شامل اطلاعات تاریخچه آموزش مدل است. این شی معمولاً از متد `fit()` مدل Keras برمی‌گردد و حاوی مقادیر دقت و خطای آموزش و اعتبارسنجی برای هر اپوک است.

    نمودارهای تولید شده به صورت زیر خواهند بود:
    - در نمودار اول، دقت آموزش و اعتبارسنجی نمایش داده می‌شود.
    - در نمودار دوم، خطای آموزش و اعتبارسنجی به نمایش در می‌آید.

    خروجی:
    - نمودارهای دقت و خطا را نمایش می‌دهد.
    """
    import matplotlib.pyplot as plt

def plot_training_history(history):
    """
    رسم گراف‌های دقت و خطا برای داده‌های آموزش و اعتبارسنجی.
    """
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]

    epochs_range = range(1, len(acc) + 1)

    plt.figure(figsize=(14, 5))

    # نمودار دقت
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label="Train Accuracy")
    plt.plot(epochs_range, val_acc, label="Validation Accuracy")
    plt.legend(loc="lower right")
    plt.title("Accuracy over Epochs")

    # نمودار خطا
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label="Train Loss")
    plt.plot(epochs_range, val_loss, label="Validation Loss")
    plt.legend(loc="upper right")
    plt.title("Loss over Epochs")

    plt.show()
```

```python
def predict_image(model, image_path, class_names, img_size=(180, 180)):
    """
    بارگذاری و پیش‌پردازش یک تصویر و پیش‌بینی کلاس آن.

    این تابع تصویر ورودی را از مسیر مشخص‌شده بارگذاری کرده، آن را پیش‌پردازش می‌کند
    (یعنی تغییر اندازه، تبدیل به آرایه و افزودن ابعاد اضافه برای پردازش به‌صورت بچ) و سپس با استفاده از
    مدل آموزش‌دیده، پیش‌بینی انجام می‌دهد. در نهایت، کلاس پیش‌بینی‌شده و میزان اطمینان پیش‌بینی را نمایش می‌دهد.

    پارامترها:
    - model: مدل شبکه عصبی آموزش‌دیده (در اینجا مدل CNN).
    - image_path: مسیر تصویر ورودی که باید پیش‌بینی شود.
    - class_names: لیستی از نام کلاس‌ها که به مدل آموزش داده شده‌اند.
    - img_size: اندازه تصویر ورودی پس از تغییر اندازه برای مطابقت با ورودی مدل (پیش‌فرض: (180, 180)).

    خروجی:
    - predicted_class: کلاس پیش‌بینی‌شده تصویر.
    - confidence: میزان اطمینان پیش‌بینی به‌صورت درصد.

    نمودار:
    - تصویر ورودی نمایش داده می‌شود و در عنوان آن، کلاس پیش‌بینی‌شده و میزان اطمینان نمایش داده می‌شود.
    """
    pass
import tensorflow as tf
import matplotlib.pyplot as plt

def predict_image(model, image_path, class_names, img_size=(150, 150)):
    """
    بارگذاری و پیش‌پردازش یک تصویر و پیش‌بینی کلاس آن.

    پارامترها:
    - model: مدل آموزش‌دیده
    - image_path: مسیر تصویر ورودی
    - class_names: لیست نام کلاس‌ها
    - img_size: اندازه تصویر ورودی (پیش‌فرض: (180,180))

    خروجی:
    - predicted_class: کلاس پیش‌بینی‌شده
    - confidence: درصد اطمینان مدل
    """
    # بارگذاری و تغییر اندازه تصویر
    img = tf.keras.utils.load_img(image_path, target_size=img_size)
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # افزودن بعد batch

    # پیش‌بینی با مدل
    predictions = model.predict(img_array, verbose=0)
    score = tf.nn.softmax(predictions[0])
    predicted_class = class_names[tf.argmax(score)]
    confidence = 100 * tf.reduce_max(score)

    # نمایش تصویر با عنوان شیک
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"{predicted_class} ({confidence:.2f}%)")
    plt.show()

    return predicted_class, confidence
```
