```python
import tensorflow as tf

# 1๏ธุชูุธู ูพุงุฑุงูุชุฑูุง
IMG_SIZE = (150, 150)   # ุงูุฏุงุฒู ุชุตูุฑ
BATCH_SIZE = 32         # ุชุนุฏุงุฏ ุชุตุงูุฑ ุฏุฑ ูุฑ batch
BASE_DIR = r"C:\Users\Mahdi\Downloads\rps_dataset-20250830T081629Z-1-001\rps_dataset"  # ูุณุฑ ุฏุชุงุณุช

# 2๏ธ ุณุงุฎุช ุฏุชุงุณุช ุขููุฒุด
train_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,    # 80% ุขููุฒุดุ 20% ุงุนุชุจุงุฑุณูุฌ
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# 3๏ธ ุณุงุฎุช ุฏุชุงุณุช ุงุนุชุจุงุฑุณูุฌ
val_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

#  ฺฺฉ ฺฉุฑุฏู
print("ุฏุชุงุณุช ุขููุฒุด:", train_ds)
print("ุฏุชุงุณุช ุงุนุชุจุงุฑุณูุฌ:", val_ds)
```

```python
# ุชุนุฑู ุชุงุจุน ูพุดโูพุฑุฏุงุฒุด ุดุงูู ูุฑูุงูโุณุงุฒ ููุงุฏุฑ ูพฺฉุณู
def preprocess(image, label):
 image = tf.cast(image, tf.float32) / 255.0  
 return image, label
# ุงุนูุงู ูพุดโูพุฑุฏุงุฒุด ุฑู ุฏุชุงุณุชโูุง
train_ds = train_ds.map(preprocess)
val_ds = val_ds.map(preprocess)
```
